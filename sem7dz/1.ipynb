```python
# 1 
class Vector:
    def __init__(self, x=0, y=0, z=0):
        
        if isinstance(x, str) and y == 0 and z == 0:
            s = x.strip().strip("{}").split(",")
            assert len(s) == 3, "Формат должен быть {x, y, z}"
            x, y, z = map(float, map(str.strip, s))

        assert all(isinstance(v, (int, float)) for v in (x, y, z)), "x, y, z должны быть числами"

        self.x = x
        self.y = y
        self.z = z

    def __abs__(self):
        return (self.x**2 + self.y**2 + self.z**2) ** 0.5

    def __add__(self, other):
        if isinstance(other, Vector):
            return Vector(self.x + other.x, self.y + other.y, self.z + other.z)
        elif isinstance(other, (int, float)):
            return Vector(self.x + other, self.y + other, self.z + other)

    def __radd__(self, other):
        return self + other

    def __sub__(self, other):
        if isinstance(other, Vector):
            return Vector(self.x - other.x, self.y - other.y, self.z - other.z)
        elif isinstance(other, (int, float)):
            return Vector(self.x - other, self.y - other, self.z - other)

    def __rsub__(self, other):
        if isinstance(other, (int, float)):
            return Vector(other - self.x, other - self.y, other - self.z)

    def __mul__(self, other):
        if isinstance(other, Vector):
            return self.x * other.x + self.y * other.y + self.z * other.z
        
        elif isinstance(other, (int, float)):
            return Vector(self.x * other, self.y * other, self.z * other)


    def cross(self, other):
        return Vector(
            self.y * other.z - self.z * other.y,
            self.z * other.x - self.x * other.z,
            self.x * other.y - self.y * other.x
        )
    
    def __rmul__(self, other):
        return self * other

    def __str__(self):
        return f'x = {self.x}, y = {self.y}, z = {self.z}'
```

```python
# 2 
points = [
    Vector("{1, 2, 3}"),
    Vector("{4, 0, 6}"),
    Vector("{-4, 5, 1}"),
    Vector("{-1, 2, 0}"),
    Vector("{-8, 3, 5}")

]

cen_x = sum(vector.x for vector in points) / len(points)
cen_y = sum(vector.y for vector in points) / len(points)
cen_z = sum(vector.z for vector in points) / len(points)

print(cen_x, cen_y, cen_z)
```

```python
#3 
n = len(points)
max_area = 0
max_triangle = None
for i in range(n - 2):
    for j in range(i+1, n-1):
        for k in range(j+1, n):
            A = points[i]
            B = points[j]
            C = points[k]
            AB = B - A
            AC = C - A
            area = 0.5 * abs(AB.cross(AC))
            if area > max_area: 
                max_area = area
                max_triangle = (A, B, C)
print(max_area)
print(max_triangle[0])
print(max_triangle[1])
print(max_triangle[2])
```
